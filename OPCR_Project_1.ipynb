Notre but étant de comprendre selon quels indicateurs le score de nutrition a été attribué, nous disposons d'indicateurs pertinents. Nos prochains objectifs se décomposeront en plusieurs étapes :

- Déterminer le taux de remplissage des différentes variables, cela nous permettra de séléctionner les indicateurs que nous garderons pour l'analyse finale
- Recercher la potentiel présence de doublons, afin de réduire le nombre de lignes et d'affiner la pertinence de la future analyse de ce Dataset


```python
def isolate100gvalues(arrayOfIndexes) :
    array_of_100g = []
    for i in range (0, len(arrayOfIndexes)) :
        if "100g" in arrayOfIndexes[i] :
            print(i ,':', arrayOfIndexes[i])
            array_of_100g.append(arrayOfIndexes[i])
    if len(array_of_100g) > 10 :
        print('\n\nIl y a', len(array_of_100g), "indicateurs pour 100g, il va falloir en séléctionner une dizaine maximum")
    else : 
        print('\n\nIl y a', len(array_of_100g), "indicateurs pour 100g, gardons les tous pour construire notre modèle")
```

```python
isolate100gvalues(data.columns)
```

```python
def filling_rate(data, max_rows=data.shape[0]):
    """
    Compute sorted completion rate and number of missing values for each variable of the argument in a descending order
    
    Args:
        data(dataset) : A dataset with variables
        
    Returns:
        A specific dataframe with completion rate for each variable. 
        Schema :
            Index(string): Variable names
            Taux de completion(float): completion rate
            Nombre de valeurs manquantes(int): Number of missing values
    """
    var_dict = {}
    
    for col in data.columns:
        var_dict[col] = []
        var_dict[col].append(col)
        var_dict[col].append(round((data[col].notna().sum()/data.shape[0])*100,2))
        var_dict[col].append(data[col].isna().sum())
    
    pd.options.display.max_rows = max_rows
        
    return pd.DataFrame.from_dict(data=var_dict, orient="index", columns = ["column_name","Taux de remplissage", "Nombre de valeurs manquantes"]).sort_values(by="Taux de remplissage", ascending=False)
```

```python
filling_rate(data)
```

Ici on remarque que le taux de remplissage est très bas à partir de la ligne "fiber_100g"

On va ensuite isoler les paramètres qui ont un taux de remplissage inférieurs à 50%

```python
taux_de_remplissage_suffisant = filling_rate(data)[filling_rate(data)['Taux de remplissage'] >= 50]

print(f'\n\n il y a {taux_de_remplissage_suffisant.shape[0]} lignes ayant un taux de remplissage satisfaisant')
```
